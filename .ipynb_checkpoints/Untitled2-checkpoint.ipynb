```python
eccgraph = graph.copy()
print(eccgraph)
```

```python
def eccrec(currrow, targetrow, count, traveled):
    traveled.append(currrow)
    print(traveled)
    count = count + 1
    print(targetrow)
    if currrow == targetrow:
        return 0
    if graph[currrow, targetrow] == 1:
        return count
    else:
        for travrow in range(graph.shape[0]):
            print("Trying %d" %travrow)
            if (graph[currrow, travrow] == 1):
                print("We can jump to %d" % travrow)
                if not travrow in traveled:
                    print("Jumping to %d" % travrow)
                    if eccrec(travrow, targetrow, count, traveled):
                        return eccrec(travrow, targetrow, count, traveled)
        return False

for x in range(graph.shape[0]):
    for y in range(graph.shape[1]):
        print("Checking %d , %d" % (x,y))
        eccgraph[x,y] = eccrec(x, targetrow=y, traveled=[])

print(eccgraph)
```
